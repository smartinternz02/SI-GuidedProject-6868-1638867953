# outliers

```python
data.describe()
```

```python
sns.boxplot(data["age"])
plt.show()
```

```python
q1=data["age"].describe()["25%"]
```

```python
q3=data["age"].describe()["75%"]
```

```python
iqr=q3-q1
iqr
```

```python
lower_bound=q1-(1.5*iqr)
lower_bound
```

```python
upper_bound=q3+(1.5*iqr)
upper_bound
```

```python
(data[data["age"] > upper_bound].shape[0]/data["age"].shape[0])*100
```

```python
data['age']=np.where(data['age']>upper_bound,upper_bound,np.where(data['age']<lower_bound,lower_bound,data['age']))
```

```python
sns.boxplot(data["age"])
plt.show()
```
